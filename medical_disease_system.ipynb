```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("medical_disease_prediction_10000.csv")

print(df.head())
```

```python
print("Shape of Dataset:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)
```

```python
# checking missing values
print(df.isnull().sum())
```

```python
print(df.describe())
```

```python
print(df["Disease"].value_counts())
sns.countplot(x="Disease", data=df)
plt.xticks(rotation=45)
plt.title("Disease Distribution")
plt.show()
```

```python
sns.histplot(df["Age"], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(df["BMI"], kde=True)
plt.title("BMI Distribution")
plt.show()
```

```python
sns.histplot(df["GlucoseLevel"], kde=True)
plt.title("Glucose Level Distribution")
plt.show()
```

```python
sns.histplot(df["BloodPressure"], kde=True)
plt.title("Blood Pressure Distribution")
plt.show()
```

```python
sns.boxplot(x="Disease", y="GlucoseLevel", data=df)
plt.xticks(rotation=45)
plt.title("Disease vs Glucose Level")
plt.show()
```

```python
sns.boxplot(x="Disease", y="BMI", data=df)
plt.xticks(rotation=45)
plt.title("Disease vs BMI")
plt.show()
```

```python
sns.boxplot(x="Disease", y="Cholesterol", data=df)
plt.xticks(rotation=45)
plt.title("Disease vs Cholesterol")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)

corr = numeric_df.corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
pd.crosstab(df["Gender"], df["Disease"]).plot(kind="bar")

plt.title("Gender vs Disease")
plt.show()
```

```python
pd.crosstab(df["Smoking"], df["Disease"]).plot(kind="bar")

plt.title("Smoking vs Disease")
plt.show()
```
